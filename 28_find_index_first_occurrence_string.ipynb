28. Find the Index of the First Occurrence in a String

Given two strings needle and haystack, return the index of the first occurrence of needle in haystack, or -1 if needle is not part of haystack.

 

Example 1:

Input: haystack = "sadbutsad", needle = "sad"
Output: 0
Explanation: "sad" occurs at index 0 and 6.
The first occurrence is at index 0, so we return 0.

Example 2:

Input: haystack = "leetcode", needle = "leeto"
Output: -1
Explanation: "leeto" did not occur in "leetcode", so we return -1.
 

Constraints:

1 <= haystack.length, needle.length <= 104
haystack and needle consist of only lowercase English characters.

This actually seems fairly easy, I just need to have a pointer to the start of the `needle` variable, another pointer to the `haystack` variable and move one step to the right for each letter in which they're equal. If there comes a point in which the specific character is not equal, I should shift the pointer of `needle` one step to the left too see if that new letter of `haystack` matches with the previous... actually that is incorrect, I should completely reset the pointer at position 0. 

```python
def strStr(haystack, needle):
    nee_pointer = 0
    for index in range(len(haystack)):
        if needle[nee_pointer] == haystack[index]:
            nee_pointer += 1

        else:
            nee_pointer = 0
            if needle[nee_pointer] == haystack[index]:
                nee_pointer += 1

        if nee_pointer == len(needle):
            print("I found the needle!")
            return index - nee_pointer + 1
    return -1
```

Indeed it seems that I have created an infinite loop... What if I solved it using a sliding window? I have a window of size `len(needle)` that just compares the needle to as much windows that the `haystack` can hold without overflowing

```python
def strStr(haystack, needle):
    left_wall = 0
    right_wall = left_wall + len(needle)
    while right_wall != len(haystack):
        if haystack[left_wall:right_wall] == needle:
            print("I have found the needle!")
            return left_wall
        left_wall += 1    
        right_wall = left_wall + len(needle)
    return -1
```
