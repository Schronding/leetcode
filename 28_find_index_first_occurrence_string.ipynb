# Optimal Solution (Knuth-Morris-Pratt / KMP)

Your intuition about your complexity was slightly off because Python's string slicing (`haystack[left_wall:right_wall] == needle`) actually takes $O(m)$ time under the hood (where $m$ is the length of `needle`). The language has to check the substring character by character. Thus, your sliding window solution and NeetCode's slice solution both run in $O(n \cdot m)$ time! Space complexity for slicing is also $O(m)$ since Python creates a new string copy for `haystack[left_wall:right_wall]`.

The absolute most optimal solution is the **Knuth-Morris-Pratt (KMP)** algorithm. 

### Logic
When a mismatch occurs, standard approaches (like slicing) "forget" all the previously matched characters and simply move the starting pointer by 1, checking everything again from scratch. 

KMP avoids this redundant work by pre-computing an **LPS (Longest Prefix Suffix)** array for the `needle`. 
The LPS array tells us: *if a mismatch happens at index $j$, what is the length of the longest prefix of the needle that is also a suffix up to this point?* 

By using this information, when we encounter a mismatch, we can just shift our `needle` pointer $j$ back to a safe point without having to reset the `haystack` pointer $i$.

### Complexity
- **Time Complexity:** $O(n + m)$ because we traverse the `needle` once to build the LPS array ($O(m)$), and we traverse the `haystack` once to find the match ($O(n)$).
- **Space Complexity:** $O(m)$ because we need to allocate the `lps` array of size `m`.

```python
def strStr_kmp(haystack: str, needle: str) -> int:
    if needle == "":
        return 0

    # Step 1: Build the LPS array (Longest Prefix Suffix)
    lps = [0] * len(needle)
    prevLPS = 0
    i = 1

    while i < len(needle):
        if needle[i] == needle[prevLPS]:
            lps[i] = prevLPS + 1
            prevLPS += 1
            i += 1
        elif prevLPS == 0:
            lps[i] = 0
            i += 1
        else:
            prevLPS = lps[prevLPS - 1]

    # Step 2: String matching using LPS
    i = 0  # haystack pointer
    j = 0  # needle pointer

    while i < len(haystack):
        if haystack[i] == needle[j]:
            i += 1
            j += 1
        else:
            if j == 0:
                i += 1
            else:
                j = lps[j - 1]

        if j == len(needle):
            return i - len(needle)

    return -1

# Let's test the KMP solution with your test cases
print(strStr_kmp(haystack_0, needle_0))  # Expected 0
print(strStr_kmp(haystack_1, needle_1))  # Expected -1
print(strStr_kmp(haystack_2, needle_2))  # Expected 4
```
