# Problem Description 58. Length of Last Word

Given a string s consisting of words and spaces, return the length of the last word in the string.

A word is a maximal substring consisting of non-space characters only.

Example 1:

Input: s = "Hello World"
Output: 5
Explanation: The last word is "World" with length 5.
Example 2:

Input: s = "   fly me   to   the moon  "
Output: 4
Explanation: The last word is "moon" with length 4.
Example 3:

Input: s = "luffy is still joyboy"
Output: 6
Explanation: The last word is "joyboy" with length 6.
 

Constraints:

1 <= s.length <= 104
s consists of only English letters and spaces ' '.
There will be at least one word in s.

# NeetCode Solution

We could go character by character, adding one to a counter variable `length`, resetting it when we encounter a space, and simply return whatever value `length` is equal to when we have finished going through the whole string. 

His solution is also $O(n)$ in the runtime and $O(1)$ in the memory. 

```python
def lengthOfLastWordNeetCode(s):
    i = len(s) - 1
    length = 0

    while s[i] == " ":
        i -= 1
    while i >= 0 and s[i] != " ":
        length += 1
        i -= 1
    return length

for string, expected_value in test_cases_dic.items():
    print(f"Algorithm Value: {lengthOfLastWordNeetCode(string)}; Expected value: {expected_value}")    
        
```

# Optimal Solutions and Pattern Analysis

The problem wants us to isolate the **last** word. This immediately points to the pattern of **Backwards Traversal**. Going efficiently from right to left allows us to find the answer without inspecting the whole string (best-case time is shorter, though worst-case is still $O(n)$ if the word is at the start).

### Addressing the Two `while` Loops 
In NeetCode's solution, the first loop **does** finish completely before the second one starts! This is a technique called **Phase Separation**. 
1. **Phase 1**: Skip trailing spaces.
2. **Phase 2**: Count the characters of the actual word until we hit a space (or the start of the string).

Separating these phases conceptually makes the code simpler to manage than keeping track of a `word_started` boolean flag inside a single loop. 

### The Pythonic Solution
In Python, string methods are implemented in highly-optimized C code. If we just want the most readable, idiomatic Python code, we can use `.split()` which handles multiple spaces for us automatically. However, this relies on built-in methods and uses $O(n)$ extra space to store the list of words.

### The Optimal Algorithmic Solution
If we want to maintain the $O(1)$ space complexity while avoiding the two while loops, we can use a `for` loop running backwards, checking if we started a word. This is very similar to your final solution, but streamlined by not keeping track of index pointers. Instead, we use a simple counter.

Complexities for this optimized algorithmic solution:
- **Time Complexity:** $O(n)$, since we iterate through the characters of the string from back to front at most once.
- **Space Complexity:** $O(1)$, since we only use a few integer variables.

```python
def lengthOfLastWordOptimal(s: str) -> int:
    length = 0
    # Phase 1: We could skip trailing spaces by looking at the string stripped of trailing spaces
    # By using `rstrip()`, we remove spaces at the end, eliminating the need for `word_started` logic
    # or the first phase loops altogether. This does return a new string though.
    # However, to be strict $O(1)$ memory without creating substrings, we can streamline your approach:
    
    for ch in reversed(s):
        if ch != ' ':
            length += 1
        elif length > 0:
            # We hit a space AFTER having counted some letters, which means the word has ended.
            return length
            
    return length

def lengthOfLastWordPythonic(s: str) -> int:
    # Python `.split()` automatically deals with multiple spaces and leading/trailing whitespace
    return len(s.split()[-1])
    
print("--- Optimal Algorithmic Solution ---")
for string, expected_value in test_cases_dic.items():
    print(f"Optimal Algorithmic Value: {lengthOfLastWordOptimal(string)}; Expected value: {expected_value}")

print("\n--- Pythonic One-Liner Solution ---")
for string, expected_value in test_cases_dic.items():
    print(f"Pythonic Value: {lengthOfLastWordPythonic(string)}; Expected value: {expected_value}")
```
