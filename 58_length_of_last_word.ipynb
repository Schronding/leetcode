# Problem Description 58. Length of Last Word

Given a string s consisting of words and spaces, return the length of the last word in the string.

A word is a maximal substring consisting of non-space characters only.

Example 1:

Input: s = "Hello World"
Output: 5
Explanation: The last word is "World" with length 5.
Example 2:

Input: s = "   fly me   to   the moon  "
Output: 4
Explanation: The last word is "moon" with length 4.
Example 3:

Input: s = "luffy is still joyboy"
Output: 6
Explanation: The last word is "joyboy" with length 6.
 

Constraints:

1 <= s.length <= 104
s consists of only English letters and spaces ' '.
There will be at least one word in s.

Besides the OnePiece reference I didn't catch I feel this problem must be easy... I could probably have the string and go backwards, ignoring all the whitespace. When there is one character that isn't a space I put a pointer in there, I continue up until I find another whitespace and put the end pointer in there. I check if that difference of the pointers (plus 1 as the arrays are zero-indexed) is greater than the largest word found yet. 

As the constraint says that there will be at least one word, if I never found a space then the largest size of the word of that string is the length of the string itself. This is independent of whether the word is large or small. 

```python
s_0 = "Hello World"
s_1 = "   fly me   to   the moon  "
s_2 = "luffy is still joyboy"

def lengthOfLastWord(s):
    beginning_pointer = 0
    end_pointer = 0
    space_character = " "
    index = 0
    longest_word_length = 0
    for character in s:
        if character != space_character: 
            beginning_pointer = index
        elif character == space_character and index > beginning_pointer:
            current_word_length = end_pointer - beginning_pointer + 1
            longest_word_length = max(longest_word_length, current_word_length)
        index += 1
    return longest_word_length

print(lengthOfLastWord(s_0))
print(lengthOfLastWord(s_1))
print(lengthOfLastWord(s_2))
```
